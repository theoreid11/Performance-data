```python
import pandas as pd
import csv
```

```python
data = pd.read_csv('user-export-3201559-2025_01_28_12_12_36.csv')
```

```python
# Convert the data into a DataFrame
df = pd.DataFrame(data)
df
```

```python

# Remove duplicates and extract the unique values from one column
df['$distinct_id']	

```

```python
comma_separated_ids = ", ".join(df['$distinct_id']).replace(" ", "")

data_comma_sep = [comma_separated_ids]
data_comma_sep
```

```python
output_file = 'output.csv'
```

```python
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Writing the comma-separated list as a single row
    writer.writerow(data_comma_sep)

print(f"Data has been written to {output_file}")
```

```python
farmingReward = pd.read_csv('farmingRewarded.csv')
farmingParticipate = pd.read_csv('FarmingParticipate.csv')
miningParticipate = pd.read_csv('miningParticipate.csv')
miningReward = pd.read_csv('miningRewarded.csv')
```

```python
farmingReward
```

```python
farmingParticipate
```

```python
participate_not_reward = farmingParticipate[~farmingParticipate['userId'].isin(farmingReward['userId'])]

# Convert to a comma-separated string
comma_separated_users = ",".join(participate_not_reward['userId'])

# Save to a file
with open("users_not_in_reward.csv", "w") as file:
    file.write(comma_separated_users)

print("File saved: users_not_in_reward.csv")
```

```python
participate_not_reward_mining = miningParticipate[~miningParticipate['userId'].isin(miningReward['userId'])]
comma_separated_users = ",".join(participate_not_reward_mining['userId'])

with open("users_not_in_reward_mining.csv", "w") as file:
    file.write(comma_separated_users)

print("File saved: users_not_in_reward_mining.csv")
```
