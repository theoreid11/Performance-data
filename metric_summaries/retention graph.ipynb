```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('retention rates 21:01.csv')
```

```python

# Ensure cohort_date is in datetime format
df['cohort_date'] = pd.to_datetime(df['cohort_date'])
# Define the start date for the last month
start_date = datetime.now() - timedelta(days=20)

# Filter the DataFrame
df= df[df['cohort_date'] >= start_date]

# Plot D1, D3, and D7 retention rates
plt.figure(figsize=(12, 6))
plt.plot(df['cohort_date'], df['d1_retention_rate'], label='D1 Retention Rate', marker='o', markersize=4)
plt.plot(df['cohort_date'], df['d3_retention_rate'], label='D3 Retention Rate', marker='o',markersize=4)
plt.plot(df['cohort_date'], df['d7_retention_rate'], label='D7 Retention Rate', marker='o',markersize=4)

# Add labels and title
plt.xlabel('Cohort Date')
plt.ylabel('Retention Rate (%)')
plt.title('Retention Rates Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
